```python
import pandas as pd
import joblib
from feature_engineering import preprocess_features
from model_utils import split_data, train_model, evaluate_model, save_model

df = pd.read_csv("cleaned_student_mat.csv")
X, y, preprocessor = preprocess_features(df)
```

```python
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
save_model(clf, "student_performance_model.pkl")
print("Model saved.")
```
